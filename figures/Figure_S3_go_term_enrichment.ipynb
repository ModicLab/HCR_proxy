**Create GO ontology graph**

```python
obo_file = "/ceph/hpc/home/novljanj/data_storage/projects/nucleolus_enriched_proteins/data/ontology_files/go-basic.obo"

def load_go_hierarchy(obo_path):
    graph = obonet.read_obo(obo_path)
    return graph

go_graph = load_go_hierarchy(obo_file)
print("Loaded GO graph with {} nodes and {} edges".format(len(go_graph.nodes), len(go_graph.edges)))
go_graph = go_graph.reverse()
```

**Visualise the enriched GO terms as a graph with all parent terms included**

```python
go_dag = obo_parser.GODag(obo_file)

def define_enriched_subset(df, category, num_to_show):
    df = df[df['#category'] == category]
    df.sort_values(by='false discovery rate', ascending=True, inplace=True)
    df = df[:num_to_show]
    return df

def add_head_of_ontology_to_enriched_terms(enriched_terms, ontology):
    if ontology == "GO Component":
        enriched_terms.add('GO:0005575')
    elif ontology == "GO Biological Process":
        enriched_terms.add('GO:0008150')
    elif ontology == "GO Molecular Function":
        enriched_terms.add('GO:0003674')
    return enriched_terms

def load_enriched_terms(df):
    return set(df["term ID"].dropna().tolist())

def normalize_values(values, vmin=0, vmax=None):
    vmin = vmin if vmin is not None else min(values)
    vmax = vmax if vmax is not None else max(values)
    norm = mcolors.Normalize(vmin=vmin, vmax=vmax)
    return norm, cm.get_cmap("Blues")

def build_go_subgraph(graph, enriched_terms, max_depth=3):
    subgraph_nodes = set(enriched_terms)

    for depth in range(1, max_depth + 1):
        new_nodes = set()
        for node in subgraph_nodes:
            if node in graph:
                new_nodes |= set(graph.predecessors(node))

        subgraph_nodes |= new_nodes
        print(f"At depth {depth}: {len(subgraph_nodes)} nodes in subgraph")

    return graph.subgraph(subgraph_nodes).copy()

def annotate_go_graph(go_graph, go_term_enrichment):
    for node in go_graph.nodes:
        if node in go_term_enrichment["term ID"].tolist():
            fdr_value = -np.log10(go_term_enrichment.loc[go_term_enrichment["term ID"] == node, "false discovery rate"].values[0])
            obs_count = np.log10(int(go_term_enrichment.loc[go_term_enrichment["term ID"] == node, "observed gene count"].values[0])) * 200
        else:
            fdr_value, obs_count = 0, 0

        go_graph.nodes[node]["fdr_value"] = fdr_value
        go_graph.nodes[node]["obs_count"] = obs_count

    return go_graph

def adjust_positions(pos, shift_amount=10):
    y_levels = {}  
    for node, (x, y) in pos.items():
        if y not in y_levels:
            y_levels[y] = []
        y_levels[y].append((node, x))

    new_pos = pos.copy()
    for y, nodes in y_levels.items():
        nodes.sort(key=lambda item: item[1]) 
        for i, (node, x) in enumerate(nodes):
            shift = shift_amount if i % 2 == 0 else -shift_amount  
            new_pos[node] = (pos[node][0], pos[node][1] + shift)

    return new_pos

def plot_go_graph(go_graph, go_dag):
    if not go_graph.nodes:
        print("No nodes to plot!")
        return

    fdr_values = [go_graph.nodes[n]["fdr_value"] for n in go_graph.nodes]
    obs_counts = [go_graph.nodes[n]["obs_count"] for n in go_graph.nodes]
    fdr_norm, cmap = normalize_values(fdr_values)
    node_colors = [cmap(fdr_norm(v)) for v in fdr_values]
    labels = {n: (go_dag[n].name if n in go_dag else n) for n in go_graph.nodes}

    fig, ax = plt.subplots(figsize=(20, 10))
    pos = nx.nx_agraph.graphviz_layout(go_graph, prog="dot")
    pos = adjust_positions(pos, shift_amount=5)

    nx.draw_networkx_nodes(
        go_graph, pos,
        node_size=obs_counts,
        node_color=node_colors,
        cmap=cmap,
        ax=ax
    )
    nx.draw_networkx_labels(
        go_graph, pos,
        labels=labels,
        font_size=7,
        font_weight="bold",
        ax=ax
    )

    right_edges = [(u, v) for u, v in go_graph.edges() if pos[v][0] > pos[u][0]]
    left_edges  = [(u, v) for u, v in go_graph.edges() if pos[v][0] <= pos[u][0]]

    nx.draw_networkx_edges(
        go_graph, pos,
        edgelist=right_edges,
        edge_color="gray",
        arrows=True,
        connectionstyle="arc3,rad=-0.1",
        ax=ax
    )

    nx.draw_networkx_edges(
        go_graph, pos,
        edgelist=left_edges,
        edge_color="gray",
        arrows=True,
        connectionstyle="arc3,rad=0.1",
        ax=ax
    )

    ax.set_title("GO Biological Process Subgraph (Filtered & Jittered)")
    ax.axis("off")
    plt.show()


ontology = "GO Component"
num_terms_to_show = 10

subset_of_enriched_terms = define_enriched_subset(go_term_enrichment, ontology, num_terms_to_show)
enriched_terms = load_enriched_terms(subset_of_enriched_terms)
enriched_terms = add_head_of_ontology_to_enriched_terms(enriched_terms, ontology)

print(f"Found {len(enriched_terms)} enriched GO terms")

subgraph = build_go_subgraph(go_graph, enriched_terms, max_depth=3)
subgraph = annotate_go_graph(subgraph, subset_of_enriched_terms)
print(f"Final subgraph has {len(subgraph.nodes)} nodes and {len(subgraph.edges)} edges")

plot_go_graph(subgraph, go_dag)
```

**Visualise the enriched GO terms as a graph with non-enriched parent terms excluded**

```python
def simplify_go_graph(go_graph, enriched_terms):

    nodes_to_remove = [node for node in go_graph.nodes if node not in enriched_terms]

    for node in nodes_to_remove:
        ancestors = list(nx.ancestors(go_graph, node))
        children = list(go_graph.successors(node))
        
        enriched_ancestors = [ancestor for ancestor in ancestors if ancestor in enriched_terms]

        if not enriched_ancestors:
            continue 
        
        closest_ancestor = min(enriched_ancestors, key=lambda anc: nx.shortest_path_length(go_graph, anc, node))

        for child in children:
            go_graph.add_edge(closest_ancestor, child)

        go_graph.remove_node(node)

    return go_graph

def prepare_for_cytoscape_visualization(subgraph):
    subgraph_names_fdr_obs = nx.DiGraph()

    for node in subgraph.nodes:
        subgraph_names_fdr_obs.add_node(
            node, 
            name=go_dag[node].name, 
            fdr_value=subgraph.nodes[node]["fdr_value"], 
            obs_count=subgraph.nodes[node]["obs_count"]
        )

    for node in subgraph.nodes:
        for child in subgraph.successors(node):
            subgraph_names_fdr_obs.add_edge(node, child)

    return subgraph_names_fdr_obs

subgraph = simplify_go_graph(subgraph, enriched_terms)
plot_go_graph(subgraph, go_dag)

subgraph_names_fdr_obs = prepare_for_cytoscape_visualization(subgraph)

#nx.write_graphml(subgraph_names_fdr_obs, "/ceph/hpc/home/novljanj/data_storage/projects/nucleolus_enriched_proteins/data/GO_enrichment_by_string/subgraph_final.graphml")
```
